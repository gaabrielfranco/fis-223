## Introdução

Os circuitos RC são circuitos onde resistores e capacitores são conectados, normalmente com uma fonte de alimentação em série com o circuito para carregá-lo. Eles tem uma ampla aplicação prática de sistemas de distribuição de potência em residências e automóveis. Alguns exemplos mais especificados são: motores, sistemas de distribuição de energia elétrica para uso comercial, industrial e doméstico, lâmpadas, aparelhos eletrodomésticos, rádio, máquinas fotográficas, entre outros.  

Abaixo a **figura 1** mostra um circuito RC onde a associação dos componentes é em série. Quando a chave está fechada em $c$ o capacitor é carregado pela _fem_ através do resistor $R$. Após o capacitor estar carregado, a chave é comutada para $d$ e o capacitor é descarregado através de $R$. Pode-se medir facilmente a diferença de potencial $V_{f}\ (=IR)$ entre os terminais do resistor para determinar a corrente i e também medir a diferença de potencial $\Delta VC\ (=\frac{q}{C})$ entre os terminais do capacitor para determinar a carga $q$ **[1]**.

![CircuitoRC](imagens\circuitoRC.jpg "Figura 1")   
\begin{align} Figura\ 1 -  Circuito\ RC,\ que\ contém\ um\ capacitor\ C\ em\ série\ com\ resistor\ R,\ e\ uma\ fonte\ de\ potencial\ V_{f}. \end{align}

Os capacitores possuem muitas aplicações que usam sua propriedade de armazenar carga e energia, por isso é de grande interesse saber como são carregados e descarregados. O simples processo de carregar e descarregar um capacitor, verifica-se uma situação na qual ocorrem variações com o tempo das correntes, das voltagens e das potências **[2]**.

### Avaliando a descarga de um capacitor
O processo de carregar o capacitor é essencial e indispensável para que o processo de descarga ocorra. Entretanto, nosso objetivo em tal prática é avaliar o processo em que o capacitor esteja descarregando.
Para analisar a situação de descarga considere que o capacitor esteja com uma carga $Q_{0}$, a seguir removemos a bateria do circuito RC e conectamos a chave no ponto $d$. A partir de então damos partida para o cronômetro em $t=0$, nesse instante $q = Q_{0}$. Assim, o capacitor se descarrega através do resistor e sua carga diminui até $0$ **[2]**.

#### Obtenção das expressões de descarga
Através da lei da malhas de Kirchhoff para o caso da descarga temos que: $$Ri(t)+\frac{q(t)}{C} = 0$$  
$$i(t) = \frac{dq}{dt} = - \frac{q(t)}{RC}\ \ (1)$$  
A corrente $i$ agora é negativa devido ao fato da carga positiva está deixando a placa do capacitor. No instante $t=0$, quando $q = Q_{0}$, a corrente inicial é dada por $I_{0} = -\frac{Q_{0}}{RC}$. Para determinarmos $q$ em função do tempo basta integrarmos a expressão **(1)**.  

$$\int_{Q_{0}}^{q} \frac{dq'}{q'} = -\frac{1}{RC}\int_{0}^{t} dt'$$  
$$\ln{\frac{q(t)}{Q_{0}}} = - \frac{t}{RC}$$  
$$q(t) = Q_{0}e^{-\ t\ /\ RC}\ (2)$$

![CircuitoRC](imagens\carga.jpg "Figura 2")   
\begin{align} Figura\ 2 - Carga\ q\ em\ função\ do\ tempo \end{align}

A corrente $i$ é derivada da expressão anterior:
$$i(t) = \frac{dq}{dt} = - \frac{Q_{0}}{RC}e^{-\ t\ /\ RC} = I_{0}e^{-\ t\ /\ RC}\ (3)$$  
![CircuitoRC](imagens\corrente.jpg "Figura 3")   
\begin{align} Figura\ 3 - Corrente\ i\ em\ função\ do\ tempo \end{align}  
Sabendo como a corrente varia no circuito durante a descarga no capacitor estamos aptos a encontrar como as diferenças de potencial entre o resistor **($V_{de}$)** e entre o capacitor **($V_{ef}$)** variam com o tempo, então obtemos as seguintes relações: 

$$V_{de} = \frac{-Q}{C}e^{-\ t\ /\ RC}\ (4)$$  
$$V_{ef} = \frac{Q}{C}e^{-\ t\ /\ RC}\ (5)$$

Dessas equações podemos notar que o fator $e^{-\ t\ /\ RC}$ governa a dinâmica de todas as grandezas relevantes nesse processo de descarga, indicando um decrescimento exponencial. Devido a essa importância a constante $RC$ presente nesse decrescimento é usada para descrever o circuito, sendo chamada de constante de relaxação, definida como:
$$\tau = RC\ \ (6)$$

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy
from matplotlib import pylab

plt.rcParams["figure.figsize"] = (25, 16)
plt.rcParams['font.size'] = 18.0

Resistor100k = pd.read_csv('data/Resistor 100k.csv')
Resistor100k = Resistor100k[3:31]
Resistor100k["Time"] = Resistor100k["Time"].apply(lambda x: x-0.3)

plt.figure(1)
plt.plot(Resistor100k["Time"],Resistor100k["Potential"],'o', Resistor100k["Time"],Resistor100k["Potential"])
pylab.title('Gráfico do potencial pelo tempo usando o resistor de 100k')
plt.xlabel("Tempo")
plt.ylabel("Potencial")
plt.figure(2)
slope, intercept, r_value, p_value, std_err = \
stats.linregress(Resistor100k["Time"],Resistor100k["Potential"].apply(numpy.log))
line = slope*Resistor100k["Time"]+intercept

plt.plot(Resistor100k["Time"],Resistor100k["Potential"].apply(numpy.log),'o', Resistor100k["Time"], line)
pylab.title('Gráfico do potencial pelo tempo usando o resistor de 100k (Linearizado)')
ax = plt.gca()
fig = plt.gcf()
plt.xlabel("Tempo")
plt.ylabel("Potencial")
plt.show()
```

Usando novamente o software que construiu os gráficos, podemos então encontrar as inclinações das retas por ele produzido. Igualmente podemos calcular por via da melhor reta visual as inclinações encontradas no gráfico feito no papel semi-logarítmo. Para as inclinações dos gráficos feitos computacionalmente, seguindo a equação $y = Ax + B$, encontramos os seguintes valores:

$$A_{R=100k\Omega} = -4.4039203980$$  
$$A_{R=47k\Omega} = -2.47298173141$$  
$$A_{R=22k\Omega} = -1.31523429034$$  

Para as retas desenhadas no papel, foi necessário encontrar pontos sobre a melhor reta visual para serem usados para o cálculo da inclinação. Os pontos usados para cada gráfico foram:

|$t_{inicial}\ (s)$|$V_{inicial}(V)$|$t_{final}(s)$|$V_{final}(V)$  |
|------------------|------------------|----------------|------------------|
|0.00$\pm$0.02     |5.8$\pm$0.01      |1.4$\pm$0.02    |0.0140$\pm$0.0001 |
|0.00$\pm$0.02     |5.8$\pm$0.01      |1.8$\pm$0.02    |0.0920$\pm$0.0001 |

<div align="center"><h4>Tabela 7 - Pontos usados para o cálculo da inclinação</h4></div>  

Depois disso foi aplicada função ln nos valores de potencia, e usada a fórmula $A' = \frac{\ln{(y_{final})} - \ln{(y_{inicial})}}{x_{final} - x_{inicial}}$. Os valores encontrados para a inclinação usando a melhor reta no gráfico feito a mão e este método foram:
$$A'_{R=100k\Omega} = -4.3\pm0.1$$  
$$A'_{R=47k\Omega} = -2.3\pm0.1$$  
$$A'_{R=22k\Omega} = -1.3\pm0.1$$ 

Agora considerando a equação $y = Ax + B$ em relação com a equação **(7)**, podemos dizer que $A = \frac{-1}{RC}$ e sabendo da equação **(6)** podemos dizer que: $$\tau = \frac{-1}{A}$$
Usando então os valores de A encontrados podemos calcular os valores de $\tau$ baseados nas retas desenhadas via software, e no papel semi-logarítmo. A unidade de $\tau$ pode ser encontrada sabendo a definição dele **(6)**, e que no SI a resistência tem unidade $\Omega$ (Ohm)  que é equivalente a Volt por Ampere, enquanto que a capacitância tem unidade **F** (Farad) que é equivalente a Coloumb por Volt. Como ampere é equivalente a Coulomb por segundo, podemos simplificar a multiplicação de $\Omega$ e **F** como:
$1 \Omega \times F = 1\frac{V}{A}\times\frac{C}{V} = 1\frac{C}{A} = 1\frac{C\ s}{C} = 1 s $. Esses valores encontrados estão abaixo:  

$$\tau_{R=100k\Omega} = 0.7603208093\ s$$  
$$\tau_{R=47k\Omega} = 0.4043701526\ s$$  
$$\tau_{R=22k\Omega} = 0,2270704076\ s$$  
$$\tau'_{R=100k\Omega} = 0.76\pm0.06\ s$$  
$$\tau'_{R=47k\Omega} = 0.43\pm0.01\ s$$  
$$\tau'_{R=22k\Omega} = 0.2323\pm0.0005\ s$$  

Em comparação, usando a fórmula **(6)** direto para os valores de $R$ e $C$ conhecidos do equipamento usado o 
$\tau_{esperado}$ de para cada resistor é:  
$$\tau_{esperado(100k\Omega)} = 1.00\ s$$  
$$\tau_{esperado(47k\Omega)} = 0.47\ s$$  
$$\tau_{esperado(22k\Omega)} = 0.22\ s$$  

Agora, com os dados experimentais e esperados, podemos calcular o erro relativo percentual para cada cada parte do experimento, calculados via software ou no papel, e na média dos dois. Usando a fórmula $E\% = \frac{|valor_{medido} - valor_{esperado}|}{valor_{esperado}}\times 100\% $ encontramos:

$$E\%_{R=100k\Omega} = 23.96791907\%$$  
$$E\%_{R=47k\Omega} = 6.56298474\%$$  
$$E\%_{R=47k\Omega} = 0.70704076\%$$  
$$E'\%_{R=100k\Omega} = 24\%$$  
$$E'\%_{R=47k\Omega} = 4\%$$  
$$E'\%_{R=47k\Omega} = 1.23\%$$  

## Conclusão
Pode-se notar que de fato o menor erro foi encontrado para o resistor de $22k\Omega$, o que é um pouco incomum considerando que seu $\tau$ é o menor de todos, e portanto teria a descarga mais rápida, possuindo então menos pontos medidos de potencial, e deveria dessa forma ter a reta menos precisa. Todavia, como pode ser visto mais claramente com os gráficos feitos computacionalmente, que mesmo aplicando a função $\ln$ na equação **(4)**, o resultado não é um reta perfeita, praticamente uma curva dependendo dos pontos usados. Considerando que o grupo deste trabalho repetiu este experimento varias vezes e pegou destes experimentos o conjunto de dados que geravam a melhor linearização, e mesmo assim os resistores de $47k\Omega$ e $100k\Omega$ tem um erro maior, acreditamos que talvez houvesse um defeito no equipamento ou um mal-contato, que fez com que um outro termo fosse acrescido a equação **(4)** de forma que mesmo ao aplicar a função $\ln$, não seria encontrada uma reta. De fato, como o erro da reta feita a mão, somente com 10 pontos para o resistor de $47k\Omega$ tem erro menor que o computacional, a teoria de que os pontos experimentais podem conter um erro tem um argumento válido.
