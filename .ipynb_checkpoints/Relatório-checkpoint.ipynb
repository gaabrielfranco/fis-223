# Primeiro Relatório de FIS225 - Circuíto RC

## Objetivos

O objetivo dessa atividade prática é verificar como a diferença de potencial de um capacitor varia em função do tempo, especificamente quando o mesmo é sujeito ao processo de descarga. Além disso, espera-se que a interpretação do significado do produto da resistência pela capacitância em um circuito RC, bem como sua importância em aplicações tecnologicas, sejam esclarecidas.

## Materiais

+ Placa de circuito Vernier (VCB), com chave **S**
+ Capacitor com capacitância igual a **10 $\mu$F**
+ Resistores de resistência igual a **22 k$\Omega$**, **47 k$\Omega$** e **100 k$\Omega$**
+ Interface Vernier com sensor de voltagem
+ Cinco fios com garra jacaré com isolamento
+ Fonte de alimentação de, no mínimo, **5 V**
+ Computador com o programas **Logger Pro** instalado corretamente

## Metodologia

Inicialmente foi montada a placa de circuito Vernier usando os fios com garra jacaré ligando os seguintes pontos da placa: **1-32, 33-18, 19-20, 21-35, 35-34**. A placa e os cabos com garra jacaré estão mostrados na figura 1 e um diagrama mostrando as conexões está ilustrados na figura 2.
![Placa e fios](imagens/vcb.jpg "Figura 1") \begin{align} Figura\ 1 - Placa\ de\ circuitos\ Vernier\ e\ fios\ com\ garra\ jacaré \end{align}
![Diagrama](imagens/diagram1.jpg "Figura 2") \begin{align} Figura\ 2 - Diagrama\ da\ configuração\ inicial\ do\ circuito\ para\ o\ resistor\ de\ 100\ k\Omega \end{align}

Tendo feito isso, a fonte de alimentação foi ligada no circuito nos encaixes **J1** e **J2** da placa, enquanto que o sensor de voltagem foi ligado nos pontos **20** e **21** da placa, paralelo ao capacitor de **10 $\mu$F**. Além disso, o sensor também foi ligado à interface Vernier e a mesma foi ligada ao computador.

Com tudo conectado, foi iniciado o programa Logger Pro no computador ligado à interface Vernier. A interface desse software está ilustrada na figura 3. Na interface é exibida um gráfico da voltagem em função do tempo e uma leitura instantânea da voltagem medida pelo sensor. Para poder calcular corretamente os valores de voltagem o sistema é calibrado marcando o estado atual como **0V**, através do botão com o ícone **$\varnothing$** circulado na figura 3.

Após a calibração, a fonte foi ajustada lentamente até um valor próximo de **5V**, baseado na indicação da voltagem mostrada pelo programa Logger Pro. A partir desse ponto, tendo o capacitor carregado, foi iniciado a captura de dados do sensor através da função *"Collect"* do software usado, também circulado na figura 3. Simultaneamente ao inicio da captura foi deslocada a chave S da placa Vernier para a posição de descarregamento do circuito, ou seja, para baixo. Ao fim do descarregamento do capacitor, o botão *"Stop"* foi pressionado e os dados coletados pelo programa foram exportados para serem usados para a criação dos gráficos inseridos na seção de Resultados abaixo.
![Interface do software](imagens/logger2.png "Figura 3") \begin{align} Figura\ 3 - Interface\ do\ programa\ Logger\ Pro \end{align}

Ao termino da medição da voltagem para o resistor de **100 k$\Omega$**, os fios com garra jacaré ligados aos pontos **33-18** e **19-20** foram trocados para as posições **33-16** e **17-20**, para o experimento com o resistor de **47 k$\Omega$**. A chave S foi religada, ou seja, posicionada para cima, e novamente os passos relacionados a medição da voltagem no software instalado foram repetidos. Numa última medição novamente os fios **33-16** e **17-20** foram reposicionados nas posições **33-14** e **15-20**, para fazer o experimento com o resistor de **22 k$\Omega$**.


## Resultados experimentais e Discussão

Seguindo o procedimento descrito na seção anterior obtivemos os dados descritos abaixo.

Para o capacitor de **100 k$\Omega$** os dados estão descritos na tabela 1, para o capacitor de **47 k$\Omega$** os dados estão descritos na tabela 2, e por fim, para o capacitor de **22 k$\Omega$** os dados estão descritos na tabela 3.

Juntamente com as tabelas, foram feitos 3 gráficos em uma escala linear, e mais 3 gráficos com o eixo dos valores do potencial na escala logarítmica, tornando o resultado numa reta.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy
from matplotlib import pylab

plt.rcParams["figure.figsize"] = (25, 16)
plt.rcParams['font.size'] = 18.0

Resistor22k = pd.read_csv('data/Resistor 22k.csv')
Resistor22k = Resistor22k[3:18]
print(Resistor22k)

plt.figure(1)
plt.plot(Resistor22k["Time"],Resistor22k["Potential"],'o', Resistor22k["Time"],Resistor22k["Potential"])
pylab.title('Gráfico do potencial pelo tempo usando o resistor de 22k')

plt.figure(2)
slope, intercept, r_value, p_value, std_err = stats.linregress(Resistor22k["Time"],Resistor22k["Potential"].apply(numpy.log))
line = slope*Resistor22k["Time"]+intercept

plt.plot(Resistor22k["Time"],Resistor22k["Potential"].apply(numpy.log),'o', Resistor22k["Time"], line)
pylab.title('Gráfico do potencial pelo tempo usando o resistor de 22k (Linearizado)')
ax = plt.gca()
fig = plt.gcf()

plt.show()
print("A = "+str(slope))

#Resistor22k.plot(x='Time', y='Potential', title="Gráfico do potencial pelo tempo usando o resistor de 22k")
#Resistor22k.plot(x='Time', y='Potential',marker= 's', logy=True, title="Gráfico do potencial pelo tempo usando o resistor de 22k (Linearizado)")
```
